## 1. Problem Statement
Given an integer array `nums`, move all `0`'s to the end of it while maintaining the relative order of the non-zero elements.

**Note** that you must do this in-place without making a copy of the array.

**Example:**
```text
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

## 2. Approach Overview

### ðŸ¢ Approach 1: Extra Array (Two Pass)

> ðŸ’­ **My Initial Thought:**  
> When I first saw this problem, my instinct was simple: collect all non-zero numbers into a new array, then copy them back and fill the rest with zeros. It seemed straightforwardâ€”separate the zeros from non-zeros, then reconstruct the array in the right order.

The extra array approach creates a separate list for non-zero elements, then reconstructs the original array by placing non-zeros first followed by zeros.

#### Key Idea:
1. Create a new array to store all non-zero elements.
2. Iterate through original array and collect non-zeros.
3. Copy non-zero elements back to original array starting from index 0.
4. Fill remaining positions with zeros.
5. Maintain relative order of non-zero elements.

#### Logic Breakdown:
* **Step 1:** Initialize empty array `non_zero = []`
* **Step 2:** First passâ€”collect non-zeros:
  - Iterate through each `num` in `nums`
  - If `num != 0`, append to `non_zero`
* **Step 3:** Second passâ€”copy non-zeros back:
  - For `j` from `0` to `len(non_zero) - 1`

#### ðŸ“ˆ Complexity Analysis

**Time Complexity:** $O(N)$
- ðŸŸ¡ **First Loop:** Iterate through all $N$ elements to collect non-zeros

- ðŸŸ¢ **Total:** $3N$ operations in worst case, simplified to $O(N)$

**Space Complexity:** $O(N)$

- ðŸ“¦ **Worst Case:** Input `[1,2,3,4,5]` creates array of size 5

````python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        non_zero=[]
        for num in nums:
            if num!=0:
                non_zero.append(num)
        for j in range(len(non_zero)):
            nums[j]=non_zero[j]
        for j in range(len(non_zero),len(nums)):
            nums[j]=0
        return nums
````
