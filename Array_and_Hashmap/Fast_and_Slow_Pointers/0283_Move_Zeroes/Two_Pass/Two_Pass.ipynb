## 1. Problem Statement
Given an integer array `nums`, move all `0`'s to the end of it while maintaining the relative order of the non-zero elements.

**Note** that you must do this in-place without making a copy of the array.

**Example:**
```text
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

## 2. Approach Overview

### ðŸŽ¯ Approach 2: Two Pointer (Overwrite & Fill)

> ðŸ’¡ **My Breakthrough Realization:**  
> After seeing my extra array solution waste memory, I had an insight: instead of collecting non-zeros in a separate array, what if I used a pointer to track where the next non-zero should go in the original array? I could overwrite positions as I scan, eliminating the need for extra space!

The two-pointer approach uses one pointer to scan the array and another to track the position where the next non-zero element should be placed, achieving in-place modification.

#### Key Idea:
1. Use a write pointer (`idx`) starting at position 0.
2. Use a read pointer (`i`) to scan through the array.
3. When we find a non-zero, write it at `idx` and increment `idx`.
4. After placing all non-zeros, fill remaining positions with zeros.
5. No extra array neededâ€”everything happens in-place!

#### Logic Breakdown:
* **Step 1:** Initialize write pointer `idx = 0`
* **Step 2:** First passâ€”place all non-zeros:
  - For `i` from `0` to `n-1`
  - If `nums[i] != 0`:
    - Write `nums[idx] = nums[i]`
    - Increment `idx++`
  - (Zeros are skipped, idx stays behind)
* **Step 3:** Second passâ€”fill remaining with zeros:
  - For `i` from `idx` to `n-1`
  - Set `nums[i] = 0`
* **Step 4:** Return modified array

## ðŸ’¡ Final Thoughts

**What This Approach Taught Me:**

   Keeping clear roles (read vs write) prevents bugs and makes the algorithm easy to reason about.

4. **Two Passes Can Be Better Than One:**  
   While one-pass solutions exist, the clarity of "place non-zeros, then fill zeros" often outweighs the minor efficiency gain.

````python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        idx=0
        for i in range(len(nums)):
            if nums[i]!=0:
                nums[idx]=nums[i]
                idx+=1
        for i in range(idx,len(nums)):
            nums[i]=0
        return nums
````

````python
solution=Solution()
l=[0,1,0,3,12]
solution.moveZeroes(l)
````
