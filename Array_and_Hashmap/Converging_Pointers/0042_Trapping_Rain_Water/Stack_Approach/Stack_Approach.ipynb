## 1. Problem Statement
Given `n` non-negative integers representing an elevation map where the width of each bar is `1`, compute how much water it can trap after raining.

**Example:**
```text
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. 
In this case, 6 units of rain water (blue section) are being trapped.

## 2. Approach Overview

### ðŸ“š Approach 3: Stack (Layer-by-Layer)

> ðŸ¤” **My Different Perspective:**  
> After mastering the left-right maximum approaches, I discovered a completely different way to think about this problem. What if instead of calculating water column-by-column (vertically), I calculated it layer-by-layer (horizontally)? This led me to a stack-based solution that processes water in a unique way.

The stack approach calculates trapped water horizontally by identifying "valleys" between bars and filling them layer by layer as we encounter taller bars.

#### Key Idea:
1. Use a stack to keep track of bar indices in a specific order.
2. When we find a bar taller than the stack's top, we've found a "right wall."
3. Pop the stack to get the "valley bottom."
4. The remaining stack top becomes the "left wall."
5. Calculate the water trapped in this horizontal layer between walls.
6. Continue until all layers are processed.

> ```
> Stack maintains indices in ascending height order (mostly)

> When we find taller bar:

>   2. Pop to find valley bottom
>   3. Remaining top is left boundary
>   4. Current position is right boundary
>   5. Calculate horizontal water strip
> 
> Key: We process water as we discover right walls,
>      filling from bottom to top, layer by layer
> ```

#### Logic Breakdown:
* **Step 1:** Initialize empty stack and `result = 0`
* **Step 2:** Iterate through each bar with index `i`
* **Step 3:** While stack is not empty AND current bar is taller than bar at stack top:
  - **Step 3a:** Pop `top` (this is the valley bottom)
  - **Step 3b:** If stack becomes empty, no left wall exists, break
  - **Step 3c:** Get `left_end = stack[-1]` (left boundary)
  - **Step 3d:** Calculate `width = i - left_end - 1` (distance between walls)
  - **Step 3e:** Calculate `height = min(height[i], height[left_end]) - height[top]`
  - **Step 3f:** Add `width Ã— height` to result (water in this layer)
* **Step 4:** Push current index `i` onto stack
* **Step 5:** Return total result

#### ðŸ“ˆ Complexity Analysis

**Time Complexity:** $O(N)$
- ðŸŸ¢ **Single Pass:** Each element is pushed onto stack once
- ðŸŸ¢ **Amortized Pops:** Each element is popped at most once
- ðŸŸ¢ **Total Operations:** Each of $N$ elements: 1 push + at most 1 pop = $2N$ operations
- ðŸŸ¢ **Simplified:** $O(N)$ with slightly higher constant factor than two-pointer

**Space Complexity:** $O(N)$
- ðŸ“¦ **Stack Storage:** Worst case stores all indices (when heights are strictly decreasing)
- ðŸ“¦ **Example:** `[5,4,3,2,1]` â†’ stack grows to size 5
- âš–ï¸ **Trade-off:** Uses $O(N)$ space like precomputation, but different algorithmic approach

#### âš¡ Performance Characteristics

**Real Performance:**

- Similar speed to precomputation (both $O(N)$ time and space)

#### ðŸš¨ Common Mistakes I Made

> **Mistake #1: Not Checking Empty Stack After Pop**
> ```python
> top = stack.pop()
> left_end = stack[-1]  # âŒ Crashes if stack is empty!
> ```
> **Problem:** After popping, stack might be empty (no left wall)  
> **Fix:** Add `if not stack: break` after popping

> **Mistake #2: Wrong Width Calculation**
> ```python
> width = i - left_end  # âŒ Includes the walls themselves!
> ```
> **Problem:** Should only count positions between walls  
> **Fix:** Use `i - left_end - 1` to exclude both boundaries

> **Mistake #3: Wrong Height Calculation**
> ```python
> h = min(height[i], height[left_end])  # âŒ Forgets valley bottom!
> ```
> **Problem:** Need to subtract the valley bottom height  
> **Fix:** `min(height[i], height[left_end]) - height[top]`

> **Mistake #4: Pushing Height Instead of Index**
> ```python
> stack.append(height[i])  # âŒ Lost index information!
> ```
> **Problem:** Need indices to calculate widths later  
> **Fix:** Always push index `i`, not `height[i]`

> ```

>

````python
class Solution:
    def trap(self, height: List[int]) -> int:
        stack=[]
        result=0
        for i in range(len(height)):
            while stack and height[stack[-1]]<height[i]:
                top=stack.pop()
                if not stack:
                    break
                left_end=stack[-1]
                width=i-left_end-1
                result+=width*(min(height[i],height[left_end])-height[top])
            stack.append(i)
        return result
````
