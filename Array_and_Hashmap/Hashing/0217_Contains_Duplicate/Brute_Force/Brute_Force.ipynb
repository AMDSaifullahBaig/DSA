## 1. Problem Statement
Given an integer array `nums`, return `true` if any value appears **at least twice** in the array, and return `false` if every element is distinct.

**Example 1:**
```text
Input: nums = [1,2,3,1]
Output: true
```
**Approach & Complexity:**

## 2. Approach: Brute Force
We compare every element with every other element to check for duplicates.
1. Iterate through the array with index `i` from `0` to `n`.
2. Iterate through the remaining elements with index `j` from `i+1` to `n`.
3. If `nums[i] == nums[j]`, we found a duplicate -> Return `True`.
4. If loops finish without matches -> Return `False`.

## 3. Complexity Analysis
* **Time Complexity:** $O(N^2)$
  - We use nested loops, leading to quadratic time growth.
* **Space Complexity:** $O(1)$
  - We do not use any extra data structures.

## 4. Limitations
* **Time Limit Exceeded (TLE):** This approach is too slow for large arrays (e.g., $N > 10,000$).

````python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        for i in nums:
            for j in nums[1+i:]:
                if i==j :
                    return True
        return False
````

````python
answer=Solution()
nums=[1,1,1,3,3,4,3,2,4,2]
print(answer.containsDuplicate(nums))
````
