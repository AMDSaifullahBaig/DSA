## 1. Problem Statement
Given two strings `s` and `t`, return `true` if `t` is an **anagram** of `s`, and `false` otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**
```text
Input: s = "anagram", t = "nagaram"
Output: true

## Approach & Complexity:
### Approach: Sorting (Brute Force)
Anagrams must contain the exact same characters in the exact same quantities. If we sort both strings, they should become identical.Check if the lengths of s and t are different. If so, return False.Sort string s alphabetically.Sort string t alphabetically.Compare the sorted strings. If they are equal, return True.

## Complexity Analysis 
* **Time Complexity:** $O(N log N)$
    * Sorting algorithms (like Timsort in Python) take $O(N log N)$ time.
* **Space Complexity:** $O(N)$
Sorting creates a new list of characters, requiring linear space.
## Limitations & Inefficiency:
This is slower than the optimal $O(N)$ approach because sorting is computationally expensive compared to a simple count.

````python
class Solution():
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s)!=len(t):
            return False
        return sorted(s)==sorted(t)
````

````python
s = "rat"
t = "car"
solution=Solution()
print(solution.isAnagram(s,t))
````
