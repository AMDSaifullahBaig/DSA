## 1. Problem Statement
Given two strings `s` and `t`, return `true` if `t` is an **anagram** of `s`, and `false` otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**
```text
Input: s = "anagram", t = "nagaram"
Output: true

## Approach & Complexity:

### Approach: Fixed Size Array (26 Characters)
Since the problem statement often limits the input to lowercase English letters, we can optimize space by using a fixed integer array of size 26 instead of a generic Hash Map.

1. Check if the lengths of s and t are different. If they are, return False.

2. Initialize an array count of size 26 with zeros.

3. Iterate through the strings. For each character:

    -Increment the count at the index corresponding to the character in s `(ord(char) - ord('a'))`.

    -Decrement the count at the index corresponding to the character in t.

4. After the loop, check if every value in the count array is 0. If any value is non-zero, return False.

5. Return True.

### Complexity Analysis
- ***Time Complexity: O(N)***

    * We iterate through the strings exactly once. The final check takes constant time O(26).

- ***Space Complexity: O(1)***

    * We strictly use an array of 26 integers. This is constant space regardless of the input string length.

### Limitations
- Character Set:
This approach assumes the input only contains lowercase English letters. It will not work for inputs with Unicode characters or emojis without modification (unlike the Hash Map approach).

````python
class Solution():
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s)!=len(t):
            return False
        array=[0]*26
        for i in range(len(s)):
            array[ord(s[i])-ord('a')]+=1
            array[ord(t[i])-ord('a')]-=1
        for i in array:
            if i!=0:
                return False
        return True
````

````python
s = "rat"
t = "car"
solution=Solution()
print(solution.isAnagram(s,t))
````
