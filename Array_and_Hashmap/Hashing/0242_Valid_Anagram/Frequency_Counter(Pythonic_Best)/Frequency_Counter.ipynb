## 1. Problem Statement
Given two strings `s` and `t`, return `true` if `t` is an **anagram** of `s`, and `false` otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**
```text
Input: s = "anagram", t = "nagaram"
Output: true

## Approach & Complexity:
### Approach: Pythonic Frequency Counter
We utilize Python's built-in collections.Counter class. This class automatically creates a hash map (dictionary) where keys are characters and values are their counts.
1. Check if len(s) != len(t). If so, return False.
2. Create a Counter object for s (Counter(s)).
3. Create a Counter object for t (Counter(t)).
4. Compare the two objects directly (Counter(s) == Counter(t)). Python handles the key-value comparison internally.
### Complexity Analysis
- Time Complexity: $O(N)$
    - Constructing a Counter takes linear time $O(N)$. Comparing two dictionaries also takes linear time in the worst case.
- Space Complexity: $O(1)$ (or $O(\Sigma)$)
    - The Counter stores counts for unique characters only. Since the alphabet size is constant (26 for English, or finite for Unicode), the space is considered constant relative to input size $N$
### Advantages
- Readability: This is the most concise and "Pythonic" way to solve the problem (often a one-liner).It is the most Time efficient solution in Python
- Speed: Heavily optimized C-implementation under the hood makes it very fast in practice, the Counter is implemented in CPython ,though technically same Big-O as the manual loop.

````python
from collections import Counter
class Solution():
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s)!=len(t):
            return False
        return Counter(s)==Counter(t)
````

````python
s = "rat"
t = "car"
solution=Solution()
print(solution.isAnagram(s,t))
````
