## Approach: Sorting + Two Pointers
The Two Pointer technique works efficiently on **sorted** arrays. Since the original problem asks for *indices* and sorting shifts elements around, we must track the original indices before sorting.

**Logic:**
1. **Track Indices:** Create a list of tuples `(value, original_index)` to remember where each number came from.
2. **Sort:** Sort this list based on the values.
3. **Two Pointers:** Initialize two pointers:
   - `left` at the start (`0`)
   - `right` at the end (`n - 1`)
4. **Iterate:**
   - Calculate `current_sum = nums[left] + nums[right]`
   - If `current_sum == target`: Return the original indices stored in the tuples.
   - If `current_sum < target`: We need a larger sum, so move `left` pointer to the right (`left++`).
   - If `current_sum > target`: We need a smaller sum, so move `right` pointer to the left (`right--`).

## Complexity Analysis
* **Time Complexity:** $O(N \log N)$
  - Sorting the array takes $O(N \log N)$.
  - The two-pointer pass takes $O(N)$.
  - The dominant term is $O(N \log N)$.
* **Space Complexity:** $O(N)$
  - We create a new list of tuples to store the original indices.
  - *Note: If we just needed to return the values (not indices), this could be $O(1)$ space depending on the sorting algorithm.*

## Trade-off Analysis
| Approach | Time | Space | Notes |
| :--- | :--- | :--- | :--- |
| Brute Force | $O(N^2)$ | $O(1)$ | Too slow for large inputs. |
| **Sorting + Two Pointers** | **$O(N \log N)$** | **$O(N)$** | **Good balance, but modifying the array complicates index tracking.** |
| Hash Map | $O(N)$ | $O(N)$ | Fastest, but uses extra memory. |

## Edge Case Checklist
* [ ] **Duplicate values:** Does the sort handle `[3, 3]` correctly? (Yes, separate tuples).
* [ ] **Negative numbers:** `[-3, 4]` (Sorting handles negatives correctly).
* [ ] **No solution:** The loop should finish without returning.

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        nums_sorted=[(value,idx) for idx,value in enumerate(nums)]
        nums_sorted.sort()
        left,right=0,len(nums)-1
        while (left<right):
            sum=nums_sorted[left][0]+nums_sorted[right][0]
            if sum==target:
                return nums_sorted[left][1],nums_sorted[right][1]
            elif sum>target:
                right-=1
            elif sum<target:
                left+=1
        return []
            
```

```python
answer=Solution()
nums=[2,7,11,15]
target=9
print(answer.twoSum(nums,target))
```
