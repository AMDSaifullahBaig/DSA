## 1. Problem Statement
Given an integer array `nums`, return `true` if any value appears **at least twice** in the array, and return `false` if every element is distinct.

## 2. Approach: Sorting
If an array is sorted, any duplicate elements will be positioned **adjacent** to each other (e.g., `[1, 1, 2, 3]`).
1. **Sort** the array `nums`.
2. Iterate from `0` to `n - 2`.
3. Check if `nums[i] == nums[i + 1]`.
   - If yes, return `True`.
4. Return `False` if no adjacent duplicates are found.

## 3. Complexity Analysis
* **Time Complexity:** $O(N \log N)$
  - Sorting takes $O(N \log N)$. The linear scan takes $O(N)$.
* **Space Complexity:** $O(1)$ or $O(N)$
  - Depends on the sorting algorithm. Python's Timsort uses $O(N)$ space in worst case.

## 4. Verdict
* Better than Brute Force, but generally slower than Hashing ($O(N)$).
* Useful if space is extremely limited and we can modify the input array.

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        nums.sort()
        for i in range(len(nums)-1):
            if nums[i]==nums[i+1]:
                return True
        return False
```

```python
answer=Solution()
nums=[1,1,1,3,3,4,3,2,4,2]
print(answer.containsDuplicate(nums))
```
