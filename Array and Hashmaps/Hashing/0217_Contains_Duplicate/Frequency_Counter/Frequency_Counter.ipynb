## 1. Problem Statement
Given an integer array `nums`, return `true` if any value appears **at least twice** in the array, and return `false` if every element is distinct.

## 2. Approach: Frequency Counter
We count the occurrences of every element using `collections.Counter`.
1. Create a frequency map `counts = Counter(nums)`.
2. Iterate through the `.values()` of the map.
3. If any count is `> 1`, return `True`.
4. Return `False` otherwise.

## 3. Complexity Analysis
* **Time Complexity:** $O(N)$
  - Building the counter traverses the array once.
* **Space Complexity:** $O(N)$
  - Storing counts for all unique elements.

## 4. Verdict
* Valid solution, but slightly less efficient than the standard Hash Set (early exit) because `Counter` typically processes the entire array first.

```python
from collections import Counter
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        count=Counter(nums)
        for i in count.values():
            if i>1:
                return True
        return False
```

```python
answer=Solution()
nums=[1,1,1,3,3,4,3,2,4,2]
print(answer.containsDuplicate(nums))
```
