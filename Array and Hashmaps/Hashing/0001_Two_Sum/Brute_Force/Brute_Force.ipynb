## 1. Problem Statement
Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.
* You may assume that each input would have **exactly one solution**.
* You may not use the same element twice.

**Example:**
```text
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

## Approach: Brute Force
We systematically check every possible pair of numbers to see if they sum to the target.
1. Loop through each element `i` from `0` to `n`.
2. Loop through each element `j` from `i+1` to `n`.
3. If `nums[i] + nums[j] == target`, we found the answer.

## Complexity Analysis
* **Time Complexity:** $O(N^2)$ - Due to the nested loops traversing the array.
* **Space Complexity:** $O(1)$ - No extra memory is used.

## Limitations
This approach is too slow for large inputs ($N > 10^4$). We need a faster look-up method to reduce the inner loop.

````python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i,1+len(nums)):
                if target==nums[i]+nums[j]:
                    return i,j
````

````python
answer=Solution()
nums=[2,7,11,15]
target=9
print(answer.twoSum(nums,target))
````
